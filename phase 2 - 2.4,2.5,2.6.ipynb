# 2.4 - Analysis of area vs overall crime

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('capston')
```

```python
df
```

```python
columns_to_exclude = ['STATE/UT', 'Population', 'Increase', 'Area(Km2)', 'Density',
       'Sex-Ratio', 'Literacy']
```

```python
numeric_columns = df.select_dtypes(include=[int, float])
columns_to_sum = numeric_columns.columns.difference(columns_to_exclude)
df['total_crime'] = numeric_columns[columns_to_sum].sum(axis=1)
```

```python
df.to_csv('capstonf', index=False)
```

```python
df
```

```python
dfl = df.iloc[:,[3,-1]]
```

```python
dfl
```

```python
sns.violinplot(x="Area(Km2)", y="total_crime", data=dfl,palette='rainbow')
```

```python
plt.figure(figsize=(30,15))
sns.stripplot(x="Area(Km2)", y="total_crime", data=dfl)
```
