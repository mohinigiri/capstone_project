  # 2.1 Analysis of Literacy Rate vs Total Crimes.

```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv('capston')
```

```python
df
```

```python

df.columns
```

```python
columns_to_exclude = ['STATE/UT', 'Population', 'Increase', 'Area(Km2)', 'Density',
       'Sex-Ratio', 'Literacy']
```

```python
numeric_columns = df.select_dtypes(include=[int, float])

```

```python
columns_to_sum = numeric_columns.columns.difference(columns_to_exclude)
```

```python
df['total_crime'] = numeric_columns[columns_to_sum].sum(axis=1)
```

```python
df
```

```python
# 2.1 Analysis of Literacy Rate vs Total Crimes.
```

```python
anl_1 = df[['STATE/UT','Literacy','total_crime']]
anl_1
```

```python
plt.figure(figsize=(30,14))
plt.plot(anl_1 ['total_crime'],anl_1 ['Literacy'])
plt.xlabel('total_crime')
plt.ylabel('Literacy')
plt.title('crime-&-litracy',fontsize = 60,color = "red")
```

```python
plt.figure(figsize=(30,14))
plt.plot(anl_1 [ 'STATE/UT'],anl_1 ['total_crime'],marker='o', linestyle='-')
plt.xlabel('STATE/UT')
plt.ylabel('total_crime')
plt.title('crime per state',fontsize=60, color="red")
```

```python
plt.figure(figsize=(30,14))
plt.plot(anl_1 [ 'STATE/UT'],anl_1 ['Literacy'],marker='o', linestyle='-')
plt.xlabel('STATE/UT')
plt.ylabel('tLiteracy')
plt.title(r"Literacy_rate per state $",fontsize=60, color="red")
```

```python
fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(anl_1 ['STATE/UT'],anl_1 [ 'Literacy'])
axes[0].set_title("scatter")

axes[1].step(anl_1 ['STATE/UT'],anl_1 [ 'Literacy'])
axes[1].set_title("step")

axes[2].bar(anl_1 ['STATE/UT'],anl_1 [ 'Literacy'], width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(anl_1 ['STATE/UT'],anl_1 [ 'Literacy'], color="green", alpha=0.5);
axes[3].set_title("fill_between");
```

```python
fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(anl_1 ['STATE/UT'],anl_1 [ 'total_crime'])
axes[0].set_title("scatter")

axes[1].step(anl_1 ['STATE/UT'],anl_1 [ 'total_crime'])
axes[1].set_title("step")

axes[2].bar(anl_1 ['STATE/UT'],anl_1 [ 'total_crime'], width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(anl_1 ['STATE/UT'],anl_1 [ 'total_crime'], color="green", alpha=0.5);
axes[3].set_title("fill_between");
```

```python

fig, ax1 = plt.subplots()

ax1.plot(anl_1 ['STATE/UT'],anl_1 [ 'Literacy'], color="blue")
ax1.set_ylabel('Literacy', fontsize=18, color="blue")

    
ax2 = ax1.twinx()
ax2.plot(anl_1 [ 'STATE/UT'],anl_1 ['total_crime'], color="red")
ax2.set_ylabel('total_crime', fontsize=18, color="red")
```
